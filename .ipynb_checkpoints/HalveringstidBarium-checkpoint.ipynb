# Halveringstid til Barium 137m
Vi leser inn målingene, gjør en kurvetilpassing og beregner halveringstiden fra denne modellen.
Her er starten av programmet, vi gjør det ferdig sammen.

```python
# importere bibliotek
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.optimize import fsolve

Bakgrunn = 2.5  # bakgrunnsstråling

# importere data fra fil
url =   'https://raw.githubusercontent.com/LektorB/Naturfag/refs/heads/main/Data/Barium1stg2025.txt'
Data = np.loadtxt(url, skiprows=6, delimiter='\t')

#deler opp dataene i to arrayer og korrigere for bakgrunnsstråling
Tid = Data[:,0]
Tellinger = (Data[:,1] - Bakgrunn)/10

# lage en modell for dataene
def f(x, a, b):
    return a * b**x

popt, pcov = curve_fit(f, Tid, Tellinger) # finne parametrene a og b 

# plotte dataene og modellen
plt.plot(Tid, Tellinger, 'o', label='Data')

modeltid = np.linspace(0, max(Tid)*1.1, 200)
plt.plot(modeltid, f(modeltid, *popt), label='Modell')

plt.legend()
```

```python
#Numerisk løsning av  
def  fun(x):
    return f(x,*popt)-popt[0]/2

print(f'Halveringstiden til Ba-137m er målt til {fsolve(fun,100)[0]:.1f}s')
```

```python
#utregnet halveringstid
float(-np.log(2)/np.log(popt[1]))
```
