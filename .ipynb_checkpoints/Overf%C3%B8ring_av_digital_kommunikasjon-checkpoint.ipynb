Under ser du en ganske lang kode som tar inn et tall i to-tallsystemmet og lager modulerte bølger som kan formidle dette tallet. Den eneste delen av koden som du trenger å endre er linje fire, hvor man legger inn meldingen.

## Oppgave 1
kjør koden under og studer resultatet.  
Bruk figuren, samt det du har lært til å svare på spørsmålene fra boken som er limt inn lenger nede i notatblokken.
  

```python
import numpy as np
import matplotlib.pyplot as plt

#Lag din egen kode, bruk null og en. Husk å start og slutt med fnutt '  
melding = '011001'

#Vi snur meldingen slik at venster del av meldingen kommer 
#til høyre i grafen.
melding = melding[::-1]

# Lager liste av meldingssrengen
data = [int(i) for i in melding] 

# lager funskjoner som kan bruke meldingen til å modulere en bølge
def amplitude(x,data):
    pos = int(x)
    return (1+data[pos])*np.sin(6*2*np.pi*x)

def frekvens(x,data):
    pos = int(x)
    return 2*np.sin((4+2*data[pos])*2*np.pi*x)

def fase(x,data):
    pos = int(x)
    return 2*np.sin(6*2*np.pi*x-np.pi*(1-data[pos]))

# lager en array for x-verdier og tilpasser lengden til lenden med data vi vil sende
xverdier = np.linspace(0,len(data),1000)
xverdier = xverdier[:-1]

# regner ut yverdier til de ulike måten å modulere bølger på
Bærebølge = 2*np.sin(12*np.pi*xverdier)
ASK = [amplitude(x,data) for x in xverdier]
FSK = [frekvens(x,data) for x in xverdier]
PSK = [fase(x,data) for x in xverdier]


# lager fire koordinatsystem i en figur
# Fyller et og et koordinatsystem med hver sin bølge:
fig, axs = plt.subplots(nrows=4, ncols=1, figsize=(8, 6), tight_layout=True)
axs[0].plot(xverdier,Bærebølge)
axs[0].set_xticklabels([])
axs[0].set_title('Umodulert bærebølge')

axs[1].plot(xverdier,ASK)
axs[1].set_xticks(np.arange(len(melding)))
axs[1].set_xticklabels([])
axs[1].set_title('Amplitudeskiftmodulasjon ASK')
axs[1].grid()

axs[2].plot(xverdier,FSK)
axs[2].set_xticklabels([])
axs[2].set_xticks(np.arange(len(melding)))
axs[2].set_title('Frekvensskiftmodulasjon FSK')
axs[2].grid()

axs[3].plot(xverdier,PSK)
axs[3].set_xticks(np.arange(len(melding)))
axs[3].set_title('Faseskiftmodulasjon PSK')
axs[3].grid()

plt.show()
```
