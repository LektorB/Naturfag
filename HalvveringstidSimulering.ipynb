```python
# laster in bibliotek
import numpy as np
import matplotlib.pyplot as plt

# antall terninger
n = 100
# antall øyne på terningene
m = 6
# antall kast
k = 20 

for i in range(k):
    plt.plot(i,n, 'or') # plotter antall terninger som er igjen før kastet
    terningkast = np.random.randint(1, m+1, n) # kast terningene
    antall_topp = np.sum(terningkast == m)  # teller antall terningkast som fikk toppverdi (seks på vanlig terning)
    n -= antall_topp  # antall terninger som er igjen etter kastet blir oppdatert
   

plt.xlabel('Antall kast')
plt.ylabel('Antall terninger igjen')
plt.title('terningmodell av radioaktivt henfall')
#ordner x-aksen til kun å vie hele tall
plt.xticks(np.arange(0, k+1, 5));
    
```

Forhåpentligvis ser du at svarene varierer litt fra gang til gang. Dette er fordi vi simulerer et tilfeldig forsøk. Men kanskje ser du også et mønster i svarene. Vi kan lage en matematisk modell for å forutsi det gjennomsnittlige resultatet. Vi bruker en eksponensiell modell:

$$N(t) = N_0 \cdot vekstfaktor^t$$


Når vi trillet terninger gjorde vi en kurvetilpassing til målingene våre. Denne gangen skal vi bygge den grafen vi forventer å få. Da vil man foråpentligvis se at de simulerte målingene "hopper" rundt modellen. I dette tilfellet er $N_0$ antall terninger vi starter med, $t$ er antall kast. Om vi forklarer med vanlig terninger, er $vekstfaktor$ hvor mange terninger vi forventer at skal være igjen etter hvert kast. Siden det er vi forventer at 1/6 ternignen blir fjernet, er 5/6 av terningene igjen med andre ord er $vekstfaktor = 5/6$.
