```python
# laster in bibliotek
import numpy as np
import matplotlib.pyplot as plt

# antall terninger
n = 100
# antall yne p terningene
m = 6
# antall kast
k = 20 

for i in range(k):
    plt.plot(i,n, 'or') # plotter antall terninger som er igjen fr kastet
    terningkast = np.random.randint(1, m+1, n) # kast terningene
    antall_topp = np.sum(terningkast == m)  # teller antall terningkast som fikk toppverdi (seks p vanlig terning)
    n -= antall_topp  # antall terninger som er igjen etter kastet blir oppdatert
   

plt.xlabel('Antall kast')
plt.ylabel('Antall terninger igjen')
plt.title('terningmodell av radioaktivt henfall')
#ordner x-aksen til kun  vie hele tall
plt.xticks(np.arange(0, k+1, 5));
    
```

Forhpentligvis ser du at svarene varierer litt fra gang til gang. Dette er fordi vi simulerer et tilfeldig forsk. Men kanskje ser du ogs et mnster i svarene. Vi kan lage en matematisk modell for  forutsi det gjennomsnittlige resultatet. Vi bruker en eksponensiell modell:

$$N(t) = N_0 \cdot vekstfaktor^t$$


Nr vi trillet terninger gjorde vi en kurvetilpassing til mlingene vre. Denne gangen skal vi bygge den grafen vi forventer  f. Da vil man forpentligvis se at de simulerte mlingene "hopper" rundt modellen. I dette tilfellet er $N_0$ antall terninger vi starter med, $t$ er antall kast. Om vi forklarer med vanlig terninger, er $vekstfaktor$ hvor mange terninger vi forventer at skal vre igjen etter hvert kast. Siden det er vi forventer at 1/6 ternignen blir fjernet, er 5/6 av terningene igjen med andre ord er $vekstfaktor = 5/6$.
